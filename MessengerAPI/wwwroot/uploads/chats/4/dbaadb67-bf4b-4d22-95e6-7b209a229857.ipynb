## Переменная x

## Функция y(x) = x**3 - 2*x**2 + x

```python
def y(): return x**3 - 2*x**2 + x
```

## График

```python
import numpy as np
import matplotlib.pyplot as plt
X = np.arange(-10,10.5,0.5)
Y = X**3 - 2*X**2 + X
plt.plot(X,Y)
plt.axhline(0)
plt.axvline(0)
plt.grid(True)
plt.title('График функции')
plt.show()
```

## Минимизация

```python
opt=tf.optimizers.SGD(learning_rate=0.1)
n=100
for i in range(n): opt.minimize(y,[x])
x.numpy()
```
