```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.title('Линейный график')
plt.grid(True)
plt.show()
```

```python
labels = ['Бирюзовый', 'Синий', 'Жёлтый', 'Зелёный']
x = [13, 12, 5, 18]
colors = ('#30D5C8', '#2F35E9', '#E7FF12', '#2E7B80')
fig = plt.figure(figsize =(3,4))
plt.pie(x, labels=labels, autopct='%1.1f%%', colors=colors)
plt.show()
```
